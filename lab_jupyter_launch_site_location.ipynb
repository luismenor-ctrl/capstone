```python
import numpy as np

# 1. Definir la función de cálculo de distancia (Fórmula de Haversine)
def calculate_distance(lat1, lon1, lat2, lon2):
    # Radio de la Tierra en kilómetros
    R = 6373.0
    
    lat1_rad = np.radians(lat1)
    lon1_rad = np.radians(lon1)
    lat2_rad = np.radians(lat2)
    lon2_rad = np.radians(lon2)
    
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    
    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    return R * c

# 2. Coordenadas (Modifica proximity_lat/lon si has elegido otro punto con el ratón)
launch_site_lat = 28.573255
launch_site_lon = -80.646895
proximity_lat = 28.56367
proximity_lon = -80.56844

# 3. Calcular la distancia usando la función recién definida
distance = calculate_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)

# 4. Crear un marcador para mostrar la distancia
distance_marker = folium.Marker(
    [proximity_lat, proximity_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12pt; color:#d35400;"><b>{distance:.2f} KM</b></div>',
    )
)
site_map.add_child(distance_marker)

# 5. Dibujar la línea entre el sitio y el punto objetivo
lines = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [proximity_lat, proximity_lon]], 
    weight=1
)
site_map.add_child(lines)

# 6. Mostrar el mapa interactivo
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
